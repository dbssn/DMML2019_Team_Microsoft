```
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

matplotlib.style.use(['seaborn-talk', 'seaborn-ticks', 'seaborn-whitegrid'])
```

```

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```
df_inspections = pd.read_csv("https://drive.switch.ch/index.php/s/nlURni4cnxItrbn/download")
```

```
df_inspections = df_inspections.drop(["AKA Name", "City", "State", "Location"], axis = 1)
df_inspections.head(20)
```

```
df_licenses = pd.read_csv("https://drive.switch.ch/index.php/s/HUOmc3Db1dJ9WMh/download")
```

```
df_licenses = df_licenses.drop(['CITY','STATE','LOCATION'], axis = 1)
df_licenses.head(20)
```

```
# join the 2 dataframe
df_final = df_inspections.merge(df_licenses, left_on='License #', right_on='LICENSE NUMBER').sort_values('License #', ascending=False)
df_final.head(50)
```

```
import matplotlib.pyplot as plt
from statistics import mode 

e = df_final.groupby('Facility Type')['LICENSE NUMBER'].count().reset_index(name='number of instituions per facility').sort_values('number of instituions per facility', ascending=False).head(30)
e.plot(x='Facility Type',y='number of instituions per facility', kind = 'bar')
```

```
# transform all facility type in minuscule in order to have juste one category per institution
df_final['Facility Type'] = [str(i).lower() for i in df_final['Facility Type']] 
df_final.head()
```

```
#number institution per category
Class_Facility = df_final.groupby(['Facility Type'])['LICENSE NUMBER'].count()
Class_Facility.reset_index(drop=False).sort_values('LICENSE NUMBER', ascending=False).head(50)
```

```
# count the number of facility types (before transformation 435)
print("number of facility types: ", df_final.groupby("Facility Type").count().shape[0])
```
