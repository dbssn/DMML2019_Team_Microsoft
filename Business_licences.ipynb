```
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

matplotlib.style.use(['seaborn-talk', 'seaborn-ticks', 'seaborn-whitegrid'])
```

```
df_inspections = pd.read_csv("https://drive.switch.ch/index.php/s/nlURni4cnxItrbn/download")
```

```
df_inspections.head()
```

```
df_licenses = pd.read_csv("https://drive.switch.ch/index.php/s/HUOmc3Db1dJ9WMh/download")
```

```
df_licenses.head(80)
```

```
df_final = df_inspections.merge(df_licenses, left_on='License #', right_on='LICENSE NUMBER').sort_values('License #', ascending=False)
df_final.head(50)
```

```
import matplotlib.pyplot as plt
from statistics import mode 

e = df_final.groupby('Facility Type')['Location'].count().reset_index(name='Movies per genre').sort_values('Movies per genre', ascending=False).head(30)
e.plot(x='Facility Type',y='Movies per genre', kind = 'bar')
```
